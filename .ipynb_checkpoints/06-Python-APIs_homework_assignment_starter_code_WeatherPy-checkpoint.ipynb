```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# URL for GET requests to retrieve weather data

base_url = "http://api.openweathermap.org/data/2.5/weather?q="
url = requests.get(base_url + "pisco" + "&appid=1f3c9313e2595aa2ab8556592bf263f9").json()
print(url)
```

```python
# Create an empty list to store the responses
response_json = []
cities = ["pisco", "tasiilaq", "atuona", "albany", "maloy", "fairbanks"]
# Make a request for each city

# Create an empty dataframe with all variables that is needed
#column_names = ["City", "Cloudiness", "Country", "Date", "Humidity", "Lat", "Lng", "Max Temp", "Wind Speed"]
#df = pd.DataFrame(columns = column_names)

# Create empty lists with all variables that is needed
city = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

for x in range(len(cities)):
    print(f"Making request number: {x} for ID: {cities[x]}")

    # Get one of the posts
    #print(base_url + str(cities[x]) + "&appid=1f3c9313e2595aa2ab8556592bf263f9")
    post_response = requests.get(base_url + str(cities[x]) + "&appid=1f3c9313e2595aa2ab8556592bf263f9")
    #print(post_response.json())

    # Save post's JSON- create a dataframe here and append
    response_json.append(post_response.json())
    city.append(post_response.json()['name'])
    cloudiness.append(post_response.json()['clouds']['all'])
    country.append(post_response.json()['sys']['country'])
    date.append(post_response.json()['dt'])
    humidity.append(post_response.json()['main']['humidity'])
    lat.append(post_response.json()['coord']['lat'])
    lng.append(post_response.json()['coord']['lon'])
    max_temp.append(post_response.json()['main']['temp_max'])
    wind_speed.append(post_response.json()['wind']['speed'])
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#convert raw data to data frame
column_names = ["City", "Cloudiness", "Country", "Date", "Humidity", "Lat", "Lng", "Max Temp", "Wind Speed"]
df = pd.DataFrame(columns = column_names)
df["City"] = city
df["Cloudiness"] = cloudiness
df["Country"] = country
df["Date"] = date
df["Humidity"] = humidity
df["Lat"] = lat
df["Lng"] = lng
df["Max Temp"] = max_temp
df["Wind Speed"] = wind_speed
df
```

```python
#Write to csv
df.to_csv('cityweather.csv', index=False)
```
